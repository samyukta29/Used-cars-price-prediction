```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
s=Service(r"C:\Users\user\Desktop\Chrome\chromedriver_win32\chromedriver.exe")
driver = webdriver.Chrome(service=s)
url="https://www.cardekho.com/"
driver.get(url)
time.sleep(3)
```

```python
# input the keywords that user wants to search
user_input = input('Enter the location : ')

#finding element for search bar
search_bar = driver.find_element('ID','cardekhosearchtext')
search_bar.send_keys(user_input)
time.sleep(1)
```

```python
search = driver.find_element(By.XPATH,'searchbtn')
search.click()
```

```python
urls = []
for _ in range(50):
        time.sleep(1)
        driver.execute_script("window.scrollBy(0,40)")
```

```python
urls = []
url = driver.find_elements(By.XPATH,"//div[@class='gsc_col-xs-7  carsName']//a")
for i in url:
        urls.append(i.get_attribute('href'))
urls
```

```python
len(urls)
```

```python
car_brand =[]
make_year = []
reg_year = []
fuel = []
kms_driven = []
engin_displacement = []
no_of_owners = []
transmission = []
insurance_type = []
car_price = []
mileage=[]
max_power=[]
```

```python
for i in urls:
    driver.get(i)
    time.sleep(2)
    # to get make year of the cars
    try:
        year=driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[1]/div/div[2]")
        for i in year:
            make_year.append(i.text)
    except NoSuchElementException:#handling no such element exception
            make_year.append('No details available')
            time.sleep(2)


    # get registered year of the cars    
    try:
        reg=driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[2]/div/div[2]")
        for i in reg:
            reg_year.append(i.text)
    except NoSuchElementException:#handling no such element exception
            reg_year.append('No details available')
            time.sleep(2)
            
     # get fule type used in the cars    
    try:
        ful=driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[3]/div/div[2]")
        for i in ful:
            fuel.append(i.text)
    except NoSuchElementException:#handling no such element exception
            fuel.append('No details available')
            time.sleep(2)
            
     # get kms run   
    try:
        kms=driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[4]/div/div[2]")
        for i in kms:
            kms_driven.append(i.text)
    except NoSuchElementException:#handling no such element exception
            kms_driven.append('No details available')
            time.sleep(2)

            
      # get engin displacement in the cars    
    try:
        dis=driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[5]/div/div[2]")
        for i in dis:
            engin_displacement.append(i.text)
    except NoSuchElementException:#handling no such element exception
            engin_displacement.append('No details available')
            time.sleep(2)

            
     # get no. of owners of the cars    
    try:
        owner=driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[6]/div/div[2]")
        for i in owner:
            no_of_owners.append(i.text)
    except NoSuchElementException:#handling no such element exception
            no_of_owners.append('No details available')
            time.sleep(2)
            
            
     # get transmission details  of the cars    
    try:
        tran=driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[8]/div/div[2]")
        for i in tran:
            transmission.append(i.text)
    except NoSuchElementException:#handling no such element exception
            transmission.append('No details available')
            time.sleep(2)
            
     # get insurance details  of the cars    
    try:
        insu=driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[9]/div/div[2]")
        for i in insu:
            insurance_type.append(i.text)
    except NoSuchElementException:#handling no such element exception
            insurance_type.append('No details available')
            time.sleep(2)
            
     
            
     # get brand name of the car   
    try:
        brand=driver.find_elements(By.XPATH,"//div[@class='gsc_col-xs-12']//h1")
        for i in brand:
            car_brand.append(i.text)
    except NoSuchElementException:#handling no such element exception
            car_brand.append('No details available')
            time.sleep(2)
            
            
            
     # get brand name of the car   
    try:
        price=driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[2]/div[1]/div/div/div/div[1]/div[2]/div/div[1]/span[2]")
        for i in price:
            car_price.append(i.text)
    except NoSuchElementException:#handling no such element exception
            car_price.append('No details available')
            time.sleep(2)
```

```python
print(len(car_brand),len(make_year),len(reg_year),len(fuel),len(kms_driven),len(engin_displacement),
      len(no_of_owners),len(transmission),len(insurance_type),len(car_price))
```
